```python
import boto3
from botocore.client import Config
import os

# Configurer des informations d'identification factices
boto3.setup_default_session(
    aws_access_key_id='fake_access_key',
    aws_secret_access_key='fake_secret_key'
)

# Configurer le client S3 avec l'URL de point de terminaison
s3 = boto3.client(
    's3',
    endpoint_url='https://data.source.coop',
    config=Config(signature_version='s3v4')
)

# Fonction pour télécharger l'ensemble du dépôt
def download_directory(bucket_name, prefix, local):
    paginator = s3.get_paginator('list_objects_v2')

    for result in paginator.paginate(Bucket=bucket_name, Prefix=prefix):
        for obj in result.get('Contents', []):
            # Chemin local du fichier
            dest_pathname = os.path.join(local, os.path.relpath(obj['Key'], prefix))
            # Crée les répertoires s'ils n'existent pas
            os.makedirs(os.path.dirname(dest_pathname), exist_ok=True)
            # Téléchargement du fichier
            s3.download_file(bucket_name, obj['Key'], dest_pathname)
            print(f"Téléchargé : {dest_pathname}")

# Appel de la fonction pour télécharger tous les fichiers
download_directory('esa', 'sen12flood/labels/', './sen12flood')
```
